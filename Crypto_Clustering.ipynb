```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
df_scaled = StandardScaler().fit_transform(df_market_data)
df_scaled
```

```python
# Create a DataFrame with the scaled data
df_transformed = pd.DataFrame(df_scaled, columns=['price_change_percentage_24h', 'price_change_percentage_7d', 'price_change_percentage_14d', 'price_change_percentage_30d', 'price_change_percentage_60d', 'price_change_percentage_200d', 'price_change_percentage_1y'])

# Copy the crypto names from the original data
coin_names = df_market_data.index

# Set the coinid column as index
df_transformed['coin_id'] = coin_names
df_transformed.set_index('coin_id', inplace=True)

# Display sample data
df_transformed.head(10)
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,12))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
for i in k:
    # Inside the loop:
    # 1. Create a KMeans model using the loop counter for the n_clusters
    k_model = KMeans(n_clusters=i, random_state=1, n_init=10)
    # 2. Fit the model to the data using `df_market_data_scaled`
    k_model.fit(df_transformed)
    # 3. Append the model.inertia_ to the inertia list
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {'k': k, 'inertia': inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
df_elbow
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(
    x="k",
    y="inertia",
    title= "Elbow Curve",
    xticks=k
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** The optimal value for k appears to be k=4, with the largest "elbow" occuring there.

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled data
model.fit(df_transformed)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
k_4 = model.predict(df_transformed)

# Print the resulting array of cluster values.
print(k_4)
```

```python
# Create a copy of the DataFrame
df_predicted = df_transformed.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_predicted['cluster'] = k_4

# Display sample data
df_predicted.head()
```

```python
# Create a scatter plot using hvPlot by setting
df_predicted.hvplot.scatter(
    # `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
    x = 'price_change_percentage_24h',
    y = 'price_change_percentage_7d',
    # Color the graph points with the labels found using K-Means and 
    # add the crypto name in the `hover_cols` parameter to identify 
    # the cryptocurrency represented by each data point.
    by = 'cluster',
    hover_cols = 'coin_id'
)
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to three principal components.
clusters_pca = pca.fit_transform(df_predicted)

# View the first five rows of the DataFrame. 
clusters_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** Total variance is 0.34871677 + 0.31363391 + 0.22627118 = 0.88862186.

```python
# Create a new DataFrame with the PCA data.
df_pca = pd.DataFrame(clusters_pca, columns=['PCA1','PCA2','PCA3'])

# Copy the crypto names from the original data
df_pca['coin_id'] = df_predicted.index

# Set the coinid column as index
df_pca.set_index('coin_id', inplace=True)

# Display sample data
df_pca.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,12))
```

```python
# Create an empty list to store the inertia values
inertia2 = []

# Create a for loop to compute the inertia with each possible value of k
for i in k:
    # Inside the loop:
    # 1. Create a KMeans model using the loop counter for the n_clusters
    k_model = KMeans(n_clusters=i, random_state=1, n_init=10)
    # 2. Fit the model to the data using `df_market_data_pca`
    k_model.fit(df_pca)
    # 3. Append the model.inertia_ to the inertia list
    inertia2.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data2 = {'k': k, 'inertia': inertia2}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow2 = pd.DataFrame(elbow_data2)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_elbow2.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** The optimal value for k is 4 when using PCA.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the PCA data
model.fit(df_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
k_4_PCA = model.predict(df_pca)

# Print the resulting array of cluster values.
print(k_4_PCA)
```

```python
# Create a copy of the DataFrame with the PCA data
df_pca_predicted = df_pca.copy()

# Add a new column to the DataFrame with the predicted clusters
df_pca_predicted['cluster'] = k_4_PCA

# Display sample data
df_pca_predicted.head()
```

```python
# Create a scatter plot using hvPlot by setting
df_pca_predicted.hvplot.scatter(
    # `x="PC1"` and `y="PC2"`. 
    x="PCA1",
    y="PCA2",
    # Color the graph points with the labels found using K-Means and 
    # add the crypto name in the `hover_cols` parameter to identify 
    # the cryptocurrency represented by each data point.
    by = 'cluster',
    hover_cols='coin_id'
)
```

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.

```python
# Composite plot to contrast the Elbow curves
df_elbow.hvplot.line(
    x="k",
    y="inertia",
    title="Elbow Curves",
    label="Original",
    xticks=k
) * df_elbow2.hvplot.line(
    x="k", 
    y="inertia",
    label="PCA",
    xticks=k
)
```

```python
# Composite plot to contrast the clusters
df_predicted.hvplot.scatter(
    x='price_change_percentage_24h',
    y='price_change_percentage_7d',
    by='cluster',
    label='Original',
    hover_cols='coin_id'
) + df_pca_predicted.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by='cluster',
    label='PCA',
    hover_cols='coin_id'
)
```
